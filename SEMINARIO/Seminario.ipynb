**¿Cual es la función objetivo?** <br>
La respuesta a esta pregunta es un poco compleja ya que no se me ocurre como escribir la función objetivo con notación matemática. Un burdo intento puede ser el siguiente:
\begin{equation*}
f = min \sum_{jornada}\sum_{\forall actores \in jornada}
\end{equation*}
Aunque no sea correcta la fórmula, lo que pretendo indicar con ella es que la función objetivo es aquella que minimiza el valor de la suma de costes de jornadas, donde el coste de cada jornada está determinado por la unión de los actores que participan en la jornada.<br>


```
# calcula el coste de una jornada por las escenas que contiene
def coste_jornada(jornada, doblaje):
  cobra_jornada = []
  for actor in range(len(doblaje[0])):
    cobra_actor = 0
    for k in range(len(jornada)):
      # print(jornada,'-->',jornada[k],'-',actor,'-',doblaje[jornada[k]][actor])
      if doblaje[jornada[k]][actor] == 1:
        cobra_actor = 1
        break
    cobra_jornada += [cobra_actor]
  coste_jornadas = sum(cobra_jornada)
  return coste_jornadas

# calcula el coste de la solución
def coste_total(solucion, doblaje):
  coste_total = 0
  coste_jornadas = []
  for jornada in solucion:
    coste_jornadas += [coste_jornada(jornada, doblaje)]
  coste_total = sum(coste_jornadas)
  return coste_total
```

Como bién dice el enunciado, hay que conseguir minimizar el coste del doblaje de la película, por lo tento es evidente que se trata de un problema de minimización.

**Diseña un algoritmo para resolver el problema por fuerza bruta**

```
### FUERZA BRUTA ###

from itertools import combinations 

a=12  # con 30 actos es imposible obtener ningún resultado, por ese motivo se ha reducido a 9 actos repartidos en tres actos al día
b=3

# se generan todas las combinaciones posibles incluidas las absurdas
secuencia = []
for i in range(a):
  secuencia += [i]
comb = list(combinations(secuencia, b))  # ~> n!/p!(n-p)! combinaciones
comb2 = combinations(comb,int(a/b))      # ~> (n!/p!(n-p)!)!/(n/p)!((n!/p!(n-p)!)-(n/p))! combinaciones
# se han generado un numero ingente de posibilidades a explorar, del orden de n!!

# se inicia el contador y las variables de control
mm=0
mejor_solucion = []
mejor_coste = 10e100

# para cada combinación, se analiza si la misma es válida
for i in comb2:
  descartar = False # para saber si se va a descartar
  # para cada jornada de la posible solución
  for h in range(len(i)): 
    resto=[] # para comparar los actos de la jornada actual con las demás
    # para cada jornada de la posible solución
    for j in range(len(i)):
      # si son jornadas posteriores a la de estudio se añade a la variable de comparación
      if h > j: resto += i[j]
    # para cada escena de la jornada
    for k in i[h]:
      # si la escena ya se encuentra en otra jornada se descarta la solución
      if k in resto:
        descartar = True
        break
  # si no se ha descartado
  if not descartar: 
    mm+=1 # suma uno al contador
    # se calcula el coste de la solución
    coste_actual = coste_total(i, doblaje)
    print(coste_actual, i)
    # si se mejora la solución, se almacena
    if coste_actual < mejor_coste:
      mejor_solucion = i
      mejor_coste = coste_actual 

print('La mejor solucion encontrada es:', mejor_coste, mejor_solucion)
```

##Funciones auxiliares

```
import random
import math
import numpy as np

# Algunas de las funciones que se utilizan en el seminario vienen de la actividad guiada 3 directamente,
# otras han sufrido algunas modificaciones y otras son nuevas

# Para crear todos los Nudos y Aristas del problema
def define_Nodos_y_Aristas(doblaje):
  Nodos = []
  Aristas = []
  for i in range(len(doblaje)):
    Nodos += [i]
    for j in range(len(doblaje)):
      Aristas += [(i, j)]
  return Nodos, Aristas

# Para crear una solución inicial aleatoria
def crear_solucion(Nodos, miembros):
  escena_elegida = []
  for i in range(len(Nodos)): escena_elegida += [random.choice(list(set(Nodos) - set(escena_elegida)))]
  solucion = [escena_elegida[i:i + miembros] for i in range(0, len(escena_elegida), miembros)]
  for _ in solucion: _.sort()
  solucion.sort()
  return solucion

#calcula el cambio de coste de una jornada al intercambiar una escena
def intercambio(jornada, a, b, doblaje):
  coste_ini = coste_jornada(jornada, doblaje)
  jornada_aux = jornada.copy()
  jornada_aux[jornada_aux.index(a)] = b
  coste_fin = coste_jornada(jornada_aux, doblaje)
  return coste_fin - coste_ini

# corta el vector solución en grupos
def separa_jornadas(solucion, miembros):
  return [solucion[i:i + miembros] for i in range(0, len(solucion), miembros)]

# une los grupos de la solución en una cadena
def junta_jornadas(solucion):
  unido = []
  for i in solucion: unido += i
  return unido
```

La matriz de Feromonas es una matriz cuadrada de tamaño 30 x 30. para cada escena elegida (fila) en un dia, a las columnas correspondientes al resto de escenas elegidas ese mismo dia les corresponde un aumento de feromonas

###Funciones auxiliares

La función que añade el nuevo nodo es la misma que utilicé en la Actividad guiada 3, con las correcciones correspondientes para adaptarlo a este problema

```
def Add_Nodo(doblaje, H, T, miembros, alfa = 0.212, beta = 10.3):

  Nodos, Aristas = define_Nodos_y_Aristas(doblaje)
  Nodos_no_asignados = list(set(range(1, len(Nodos))) - set(H))
  jornadas = separa_jornadas(H, miembros)
  ultimo_nodo = jornadas[len(jornadas)-1]
  #print(ultimo_nodo)
  # si la jornada ya está completa se añade la primera escena no seleccionada a la siguiente jornada
  # si no está completa, se estudia cual añade el menor coste a la jornada
  if len(ultimo_nodo) == miembros:
    Nuevo_Nodo = list(set(range(1, len(Nodos))) - set(H))[0]
  else:
    # función de probabilidad
    estimador = []  # <--- almacenará el sumatorio del denominador de la función de probabilidad para cada arista
    # para cada nodo no asignado se calculan los parámetros de sus aristas
    for i in Nodos_no_asignados:
      tau = T[ultimo_nodo[0]][i] ** alfa
      nu = (1/(coste_jornada(ultimo_nodo + [i], doblaje)) ** beta)
      estimador += [nu * tau]

    # Con los valores calculados, se calculan los valores de la función de probabilidad
    funcion_prob = []
    # para cada valor de los términos calculado
    for valor in estimador:
      # el valor del término entre la suma de todos los términos
      funcion_prob += [valor/sum(estimador)]
    # se genera un número aleatorio entre 0 y 1
    # esto servirá como límite de elección del Nodo 
    valor = random.random() #* sum(termino)

    # se selecciona el valor de i en cuanto la suma de todos los valores 
    # anteriores supera el valor aleatorio previamente calculado
    acumulado = 0
    for i in range(len(funcion_prob)):
      acumulado += funcion_prob[i]
      # cuando la función de densidad alcanza el valor aleatorio, se selecciona ese nodo.
      if valor < acumulado: break

    Nuevo_Nodo = Nodos_no_asignados[i]
    # se devuelve el nodo con el índice obtenido previamente
  return Nuevo_Nodo

def Incrementa_Feromona(doblaje, T, H, feromonas):
  # Incrementa segun la calidad de la solución. Añadir una cantidad inversamente proporcional a la distancia total
  for i in range(len(H)):
    for k in H[i]:
      for j in list(set(H[i]) - {k}): 
        T[k][j] += feromonas * 1000 / coste_total(H, doblaje)
    # print(i[0],'-',T[i[0]])
  return T

def Evaporar_Feromonas(doblaje, T, evaporador):
  # Evapora 0.3 el valor de la feromona, sin que baje de 1
  # Mejora: Podemos ejegir diferentes funciones de evaporación dependiendo de la cantidad
  # actual y de la suma total de feromonas depositadas...
  Nodos, Aristas = define_Nodos_y_Aristas(doblaje)
  T = [[max(T[j][i] * evaporador, 1) for i in range(len(Nodos))] for j in range(len(Nodos))]
  return T
```

```
# COLONIA DE HORMIGAS

def hormigas(doblaje, N, miembros, alfa=1, beta=1, feromonas = 1, evaporador = 0.5):
  #doblaje = datos del problema
  #N = numero de iteraciones
  #miembros = escenas por dia

  # Nodos y Aristas
  Nodos, Aristas = define_Nodos_y_Aristas(doblaje)

  # Inicializa las aristas con una cantidad inicial de feromonas: 1
  # Mejora: inicializar con valores diferentes dependiendo diferentes criterios
  T = [[1 for _ in range(len(Nodos))] for _ in range(len(Nodos))]
  
  # Se generan los agentes (hormigas) que serán estructuras de caminos desde 0
  Hormiga = [[] for _ in range(N)]

  # Recorre cada agente construyendo la solución
  for h in range(N):
    Hormiga[h] = [Nodos[0]] # se elige la primera escena de la primera jornada
    # Para cada agente se construye un camino
    for i in range(len(Nodos)-1):
      #Elige el siguiente nodo
      # print(Hormiga[h])
      Nuevo_Nodo = Add_Nodo(doblaje, Hormiga[h], T, miembros) #, alfa, beta)
      Hormiga[h].append(Nuevo_Nodo)
    # Incrementa feromonas en esa arista
    T = Incrementa_Feromona(doblaje, T, separa_jornadas(Hormiga[h], miembros), feromonas)
    # print('Feromonas(1)', T)
    
    #Evapora Feromonas
    T = Evaporar_Feromonas(doblaje, T, evaporador)
    # print('Feromonas(2)', T)
    # Seleccionamos el mejor agente

  mejor_solucion = []
  mejor_coste = 10e100
  
  for h in range(N):
    Hormiga[h] = separa_jornadas(Hormiga[h],miembros)   # se divide el vector en jornadas
    for _ in Hormiga[h]: _.sort()   # se ordena cada jornada
    Hormiga[h].sort()   # se ordenan todas las jornadas
    coste_actual = coste_total(Hormiga[h], doblaje)   # se calcula el coste
    # print(coste_actual,Hormiga[h])
    if coste_actual < mejor_coste:
      mejor_solucion = Hormiga[h]
      mejor_coste = coste_actual 
  # print('La mejor solucion encontrada es:', mejor_coste, mejor_solucion)
  # print(T)
  # print('')
  return mejor_coste, mejor_solucion
```

###Funciones auxiliares

```

# Homólogo al mismo generador de vecinos de la Actividad Giada 3

def genera_vecina_aleatorio(solucion):
  # Generador de 1 solucion vecina. Intercambia 2 nodos aleatorios
  # Se eligen aleatoriamente dos jornadas de la solución
  i = random.choice(range(0, len(solucion)))
  j = random.choice(list(set(range(0, len(solucion))) - {i}))
  if j < i:
    i, j = j, i

  # La principal deferencia respecto a la AG3 es que en esta ocasión hay que intercambiar escenas entre jornadas
  jornada_A = solucion[i].copy()
  jornada_B = solucion[j].copy()
  
  # se elige aleatoriamente una toma de cada jornada
  toma_A = jornada_A[random.choice(range(0, len(jornada_A)))] 
  toma_B = jornada_B[random.choice(range(0, len(jornada_B)))]
  
  # se intercambian las tomas entre las jornadas
  jornada_A[jornada_A.index(toma_A)] = toma_B
  jornada_B[jornada_B.index(toma_B)] = toma_A

  # se construye la nueva solución vecina aleatoria
  vecina = list(solucion[:i] + [jornada_A] + solucion[i + 1:j] + [jornada_B] + solucion[j + 1:])

  # el valor booleano se utiliza para controlar el método de elección del siguiente 
  # vecino. Se ha incluido de cara a la solución final del problema y se explicará más adelante
  return True, vecina

# Función de probabilidad para determinar si se cambia a una solución peor respecto a la referencia (exponencial)
def probabilidad(T, d):
  if random.random() <= math.exp(-1*d / T):
    return True 
  else:
    return False 

def bajar_temperatura_exp(T):
  return T * .9999   # descenso exponencial

def bajar_temperatura_bolt(T, k):
  return T / (1 + math.log10(k))   # Boltzmann

def bajar_temperatura_cauchy(T, k):
  return T / (1 + k)   # Cauchy
```

### Método

```
# RECOCIDO SIMULADO

def recocido_simulado(doblaje, miembros, TEMPERATURA = 100, objetivo = 1):
  # doblaje = datos del problema
  # miembros = cantidad de tomas que se van a realizar en cada jornada
  # TEMPERATURA = valor inicial de la temperatura de recocido
  # objetivo = valor de solución -> condición de parada
  
  # Para iniciar el sistema se utiliza una solución aleatoria. Esto conduce a valores 
  # alejados del objetivo pero incluso así se llega rápidamente a un mínimo local
  Nodos, Aristas = define_Nodos_y_Aristas(doblaje)
  solucion_referencia = crear_solucion(Nodos, miembros)

  # esta variable sirve para controlar el proceso de optimización
  coste_referencia = coste_total(solucion_referencia, doblaje)

  # se inicializan las variables del resultado
  mejor_solucion = solucion_referencia
  mejor_coste = coste_referencia   

  T = TEMPERATURA  # temperatura del recocido
  siguiente_cercano = True # controla el algoritmo de selección de vecinos. No tiene utilidad 
                           # en este método, pero si en la solución final. Se explicará más adelante
  
  # Mientras se dan las condiciones, se ejecuta el bucle:
  #    · Temperatura mínima del recocido no alcanzada
  #    · coste objetivo no conseguido
  k=1 # contador de ciclos
  while T > 1 and mejor_coste > objetivo:
    # Genera una solución vecina aleatoria
    siguiente_cercano, vecina = genera_vecina_aleatorio(solucion_referencia)

    #Calcula el coste de la nueva solución
    coste_vecina = coste_total(vecina, doblaje)

    #Si es la mejor solución de todas se guarda
    if coste_vecina < mejor_coste:
      for _ in vecina: _.sort() # Se ordenan los actos dentro de las jornadas y ...
      vecina.sort()             # ... acto eguido las jornadas 
      mejor_solucion = vecina 
      mejor_coste = coste_vecina
      # print(k,'iteraciones')
      print('\t****Iteracion',k,'| Encontrada mejor solución:', mejor_coste, mejor_solucion, '\n')

    #Si la nueva vecina es mejor se cambia y si es peor se cambia según una 
    #probabilidad dependiente de T y de |coste_referencia - coste_vecina|
    p = probabilidad(T, abs(coste_referencia - coste_vecina))
    if coste_vecina < coste_referencia or p:
      solucion_referencia = vecina 
      coste_referencia = coste_vecina

    T = bajar_temperatura_exp(T)    #  --> Esta es la que mejor resultados da
    # T = bajar_temperatura_cauchy(T, k)
    # T = bajar_temperatura_bolt(T, k + 1)

    k += 1

  print('La mejor solucion encontrada es:', mejor_coste, mejor_solucion)
  # print(k,'iteraciones')
  return mejor_solucion

recocido_simulado(doblaje, 6, TEMPERATURA = 100000)
```

###Mejora del algoritmo

```
def genera_vecina_cercano(solucion, doblaje):
  # Generador de 1 solucion vecina. Intercambia 2 nodos que mejoren la solución de alguna jornada
  # dándole prioridad a las jornadas que tienen menor coste. Minimiza los coste de jornada
  # ### TECNICA VORAZ ###

  # se calcula el coste de cada jornada de la solución y se añade el valor a un vector
  coste_jornadas = [] 
  for jornada in solucion:
    coste_jornadas += [coste_jornada(jornada, doblaje)]
  

  # se selecciona la jornada que tenga el menor coste usando el vector
  for g in range(len(coste_jornadas)):
    if coste_jornadas[g] == min(coste_jornadas):
      break
  i = g

  # la jornada que se va a optimizar se almacena en el siguiente vector para no incluirlo en 
  # los análisis de jornadas posteriores
  jornada_analizada = [g] #este vector contendrá las jornadas descartadas por no poder mejorar

  # para cada jornada de la solución, se calcula el coste de cambiar un acto de la jornada 
  # seleccionada con el resto de actos. ese calculo se almacena en los índices
  for _ in range(len(solucion)):
    # vectores con los resultados
    indice_escena_ini = []  # escena que se está estudiando
    indice_escena_fin = []  # escena con la que se está comparando
    indice_jorn_fin = []    # jornada en la que se encuentra esa escena
    indice_coste = []       # diferencia de coste que supone el cambio
    # para cada acto de la jornada seleccionada
    for x in solucion[i]:
      # para cada jornada no seleccionada hasta este momento
      for y in list(set(range(0, len(solucion))) - set(jornada_analizada)):
        # para cada acto de esta segunda jornada
        for z in solucion[y]:
          # se añaden valores a los índices
          indice_escena_ini += [x]
          indice_escena_fin += [z]
          indice_jorn_fin += [y]
          indice_coste += [intercambio(solucion[i], x, z, doblaje)]
    
    #IMPORTANTE
    # Est e booleano se utiliza para controlar el método de elección del siguiente vecino:
    #   · Si es True se volverá a utilizar este método para el siguiente vecino
    #   · Si es False se utilizará el vecino aleatorio en la siguiente iteración
    movimiento_disponible = True
    
    # Se determina si los vectores de índice están vacíos --> no hay movimientos que 
    # mejoren ninguna jornada sin empeorar alguna jornada ya optimizada 
    if not indice_coste:
      movimiento_disponible = False
      # si no hay movimientos disponibles que mejoran, se sale del análisis de selección de jornada
      break
    
    # Si para la jornada seleccionada existen movimientos de mejora, se entiende que la jornada 
    # es válida para optimizar y por lo tanto se sale del análisis de selección de jornada
    if min(indice_coste) < 0:  break

    # En caso contrario, se considera que la jornada está optimizada y por lo tanto se 
    # busca la siguiente jornada con menor coste
    nuevo_minimo = coste_jornadas.copy()
    # A la jornada analizada se le pone un coste elevado para evitar que vuelva a ser seleccionado
    for jornadas in jornada_analizada: nuevo_minimo[jornadas] = len(doblaje[0]) + 1 
    # se selecciona la siguiente jornada que tenga el menor coste
    for g in list(set(range(0, len(solucion))) - set(jornada_analizada)):
      if coste_jornadas[g] == min(nuevo_minimo):
        break
    i = g
    jornada_analizada += [g]  #se añade al vector de jornadas descartadas y se inicia de nuevo el bucle

  # se inicializa la variable que almacena el sesultado
  vecina = solucion

  # si se ha determinado que hay movimientos de mejora, se entra a seleccionar uno de 
  # los actos de mejora, en caso contrario se sale con la solución inicial
  if movimiento_disponible:      

    # Se genera una máscara para seleccionar los índices con las opciones de mejora para
    # aquellas opciones que más mejoren la jornada --> minimo de coste de intercambio
    mask = []
    for h in indice_coste:
      if h == min(indice_coste): mask += [True]
      else: mask += [False]
    mask = np.array(mask)

    # Para todas las posibilidades encontradas, se selecciona una de ellas aleatoriamente
    indice_aleatorio = random.randint(0, len(np.array(indice_coste)[mask])-1)

    # y por último se intercambian las escenas entre las jornadas
    j = np.array(indice_jorn_fin)[mask][indice_aleatorio]         # Jornada destino
    toma_A = np.array(indice_escena_ini)[mask][indice_aleatorio]  # toma de la jornada a optimizar
    toma_B = np.array(indice_escena_fin)[mask][indice_aleatorio]  # toma que mejora la jornada

    # Se contruye la nueva solución vecina
    if j < i:
      i, j = j, i
      toma_A, toma_B = toma_B, toma_A
    jornada_A = solucion[i].copy()
    jornada_B = solucion[j].copy()
    jornada_A[jornada_A.index(toma_A)] = toma_B
    jornada_B[jornada_B.index(toma_B)] = toma_A
    vecina = list(solucion[:i] + [jornada_A] + solucion[i + 1:j] + [jornada_B] + solucion[j + 1:])
  
  # ORDENAR EL RESULTADO
  # en último lugar, antes de entregar el resultado, se ordenan las escenas de menor a mallor dentro de cada jornada
  for _ in vecina: _.sort()
  # y se ordenan las jornadas teniendo en cuenta solo el la escena más pequeña de la jornada
  vecina.sort()

  return movimiento_disponible, vecina
```

Al tratarse de un método de voracidad, el primer efecto que ha tenido la mejora en el método de elección de vecinos ha sido que ha desaparecido el efecto de la temperatura en el recocido simulado ya que en pocos pasos se llega al mínimo local más cercano y no le da tiempo al sistema a "enfriarse" cuando ya ha llegado a un mínimo local.

Por este motivo se ha modificado el algoritmo de recocido simulado para quitar el código relativo a la temperatura. 

Por otro lado se ha añadido un contador de ciclos y un análisis de valor objetivo, para utilizarlos como métodos de parada

Además se ha incluido un salto aleatorio cuando la función alcanza un mínimo local si éste no es menor o igual que el objetivo.

Por último, con la idea de reducir al mínimo el número de ciclos, se ha dado como solución inicial el resultado del algoritmo de Colonia de Hormigas desarrollado antes con una cantidad de 15 hormigas. Esto consigue que las distintas iteraciones comiencen con un valor bastante aproximado.

Debido a todos estos cambios, no se puede seguir llamando a este método "Recocido simulado" siembargo hay otro nombre que se puede asemejar más a esta resolución: GRASP. Más adelante daré más explicaciones.

```
def GRASP(doblaje, miembros, N = 100, objetivo = 1):
  # doblaje = datos del problema
  # miembros = cantidad de tomas que se van a realizar en cada jornada
  # N = cantidad de bucles hasta la parada --> por defecto 100
  # objetivo = valor de solución - condición de parada
  
  # Para iniciar el sistema y aprovechando que he desarrollado un algoritmo de colonia 
  # de hormigas para intentar resolver este ejercicio, la solución inicial de referencia 
  # será el resultado de aplicar la solución de hormigas para unos pocos agentes.
  vecina = hormigas(doblaje, 15, miembros, 0.6, 0.9, 3, 0.9)[1]
  # una alternativa peor pero más simple es utilizar una solución aleatoria. Esto 
  # conduce a tardar más tiempo, pero incluso así se llega rápidamente a la solución
  # Nodos, Aristas = define_Nodos_y_Aristas(doblaje)
  # solucion_referencia = crear_solucion(Nodos, miembros)

  # esta variable sirve para controlar el proceso de optimización
  coste_referencia = coste_total(vecina, doblaje)

  # se inicializan las variables del resultado
  mejor_solucion = vecina
  mejor_coste = coste_referencia   

  k = 0            # contador de iteraciones
  siguiente_cercano = True # controla el algoritmo de selección de vecinos
  
  # Mientras se dan las condiciones, se ejecuta el bucle:
  #    · número de ciclos k menor que el límite establecido
  #    · coste objetivo no conseguido
  while k < N and mejor_coste > objetivo:
    # Genera una solución vecina que mejore alguna jornada   ### BUSQUEDA VORAZ ###
    siguiente_cercano, vecina = genera_vecina_cercano(vecina, doblaje)
    if not siguiente_cercano:
      # si no hay una solución vecina que mejore alguna jornada, generar una solución
      # vecina aleatoria  ### BUSQUEDA ALEATORIA ###
      siguiente_cercano, vecina = genera_vecina_aleatorio(vecina)

    #Calcula el coste de la nueva solución
    coste_vecina = coste_total(vecina, doblaje)

    #Si es la mejor solución de todas se guarda
    if coste_vecina < mejor_coste:
      mejor_solucion = vecina 
      mejor_coste = coste_vecina
      # print(k,'iteraciones')
      print('\t****Iteracion',k,'| Encontrada mejor solución:', mejor_coste, mejor_solucion, '\n')

    k += 1
  
  # Se ordenan los actos dentro de las jornadas y posteriormente las jornadas
  # for _ in mejor_solucion: _.sort()
  # mejor_solucion.sort()

  print('La mejor solucion encontrada es:', mejor_coste, mejor_solucion)
  # print(k,'iteraciones')
  return mejor_solucion
```

###Método GRASP
Al parecer estas técnicas así aplicadas se corresponderían con un algoritmo de tipo Greedy Randomized Adaptative Search Procedures (GRASP). Esto es así porque el algoritmo desarrollado contiene distintos algoritmos heurísticos y aleatorios que mejoran el acercamiento a la solución.

Comienza con una solución inicial basada en colonia de hormigas. Con esto se acerca a una solución aceptable.

A continuación comienza un proceso iterativo de acercamiento al mínimo mas cercano mediante un algoritmo voraz de vecinos cercanos.

Finalmente, cuando no puede mejorar más esa solución, ejecuta un intercambio aleatorio a una solición peor pero cercana (multiarranque) con el objetivo de escapar de posibles mínimos locales, de manera que permite volver a calcular todas las jornadas hasta el siguiente mínimo local. 

Con la unión de todas esas operaciones, se consigue llegar a un mínimo absoluto muy rápidamente con una eficiencia casi del 100% en menos de 1000 iteraciones del método.
